```python
import os
import numpy as np
import pandas as pd
```

```python
mel_th = 0.01171875
band_th = 0.34375
gabor_th = 0.375
```

```python
df_scores=pd.read_csv("eval_utils/results/hit_rate_AM1_July24th.csv")
```

```python
_, counts_band = np.unique(df_scores['band_score']>band_th, return_counts=True)
counts_band[1]/np.sum(counts_band)
```

```python
_, counts_gabor = np.unique(df_scores['gabor_score']>gabor_th, return_counts=True)
counts_gabor[1]/np.sum(counts_gabor)
```

```python
_, counts_mel = np.unique(df_scores['mel_score']>mel_th, return_counts=True)
counts_mel[1]/np.sum(counts_mel)
```

```python
goe_path="/Users/yasminebenhamadi/PhDv.1/Yellowhammer/eval_utils/results/July24th/AM1"
files_csv=[f for f in os.listdir(goe_path) if ".csv" in f]

goe_scores=[]
for f_csv in files_csv:
    goe_scores.extend(pd.read_csv(os.path.join(goe_path, f_csv))['max'].to_list())

goe_scores=np.array(goe_scores)

_, counts_goe = np.unique(goe_scores>0.0037258610833159, return_counts=True)
counts_goe[1]/np.sum(counts_goe)
```
